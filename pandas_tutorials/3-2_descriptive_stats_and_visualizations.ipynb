### 5.21 Set up Environment

````python
# import libraries
import pandas
import numpy
import matplotlib.pyplot
import seaborn

# magic method for inline plotting
%pylab inline
````

````python
# seaborn settings

seaborn.set_palette("colorblind", desat=.6)
seaborn.set_context(rc={"figure.figsize": (8,4)})
````

### 5.22 Load Data

````python
# load titanic set
url = 'https://raw.github.com/alfredessa/pdacookbook/master/data/titanic.csv'

titanic = pandas.read_csv(url)
````

````python
# load server data set
url = 'https://raw.github.com/alfredessa/pdacookbook/master/data/serverdata.csv'

server = pandas.read_csv(url)
````

````python
# load mlb set
url = 'https://raw.github.com/alfredessa/pdacookbook/master/data/mlbsalaries.csv'

mlb = pandas.read_csv(url)
````

#### 1: Histogram

````python
# make a histogram of titanic ages

# remove NaN with `dropna()`
matplotlib.pyplot.hist(titanic.Age.dropna(), bins=25)
````

````python
# histogram of server time

matplotlib.pyplot.hist(server.Time, bins=25, color=seaborn.desaturate("indianred",1))
````

````python
# histogram of MLB salaries

matplotlib.pyplot.hist(mlb.Salary.dropna(), bins=25)
````

#### 2: Boxplot

````python
# boxplot of titanic ages

seaborn.boxplot(titanic.Age.dropna())
````

````python
# boxplot of titanic ages, grouped by gender

# seaborn can handle NaNs, so we don't always need `dropna()`
# skews results though
seaborn.boxplot(titanic.Age, groupby=titanic.Sex, vert=False)
````

````python
# boxplot of server time

seaborn.boxplot(server.Time)
````

````python
# boxplot of mlb salaries

seaborn.boxplot(mlb.Salary.dropna())
````

#### Kernel Density Estimate (KDE) 

````python
seaborn.kdeplot(titanic.Age.dropna(), shade=True)
````

````python
seaborn.kdeplot(server.Time, shade=True)
````

````python
seaborn.kdeplot(mlb.Salary.dropna(), shade=True)
````

````python
# coombine a KDE and histogram with `distplot()`

seaborn.distplot(titanic.Age.dropna())
````

````python
seaborn.distplot(mlb.Salary.dropna())
````

#### Violin Plot

````python
# combine a KDE and a boxplot with `violinplot()`

seaborn.violinplot(titanic.Age.dropna())
````

````python
# group violin plots by gender

seaborn.violinplot(titanic.Age.dropna(), groupby=titanic.Sex)
````

````python
# violin plot the mlb salaries

seaborn.violinplot(mlb.Salary.dropna())
````

#### Cumulative Distribution Function (CDF)

````python
A quick way of looking at percentiles.
````

Use:  
```seaborn.kdeplot( val = , cumulative = True)```

````python
# CDF shows the probability that a random value from the population is greater or less than a CDF bin

# Ex: there is an 80% probability that a random titanic passenger is YOUNGER than 40

seaborn.kdeplot(titanic.Age.dropna(), cumulative=True)
````

````python
# roughly 40% of server response times are LESS than 500 ms

seaborn.kdeplot(server.Time, cumulative=True)
````

````python
# about 90% of MLB players make LESS than $1M per year

seaborn.kdeplot(mlb.Salary.dropna(), cumulative=True)
````
